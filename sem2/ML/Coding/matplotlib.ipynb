```python
# import the libraries here 
import matplotlib
#matplotlib is a low level plotting python library that serves as a visualization utility
```

```python
# checking matplotlib version
print(matplotlib.__version__)
```

```python
#most of the matplotlib utilitites lies under the pyplot submodule
import matplotlib.pyplot as plt
```

```python
#drawing line through given points
import numpy as np
x = np.array([0,4,7,9,2])
y = np.array([10,14,17,19,12])
plt.plot(x , y)#sort the x and plots
plt.show( )
```

```python
#drawing line through given points
x = np.array([0,4,7,9,2])
y = np.array([10,14,17,19,12])
plt.plot(x , y, marker='o')# @every point shows dont 
plt.show( )
```

```python
#locate the given points with dots
x = np.array([0,4,7,9])
y = np.array([10,14,17,19])
plt.plot(x, y, 'o')
plt.show( )
```

```python
#format of string in the plot function is marker|line|color
#marker 'o','*','.'
#line '-',':','--','-.'
#color 'r','g','b'
x = np.array([1,2,3,4])
y = np.array([5,1,7,3])
plt.plot(x,y,'o:b')
plt.show()
```

```python
#ms - marker size
x = np.array([1,2,3,4])
y = np.array([5,1,7,3])
plt.plot(x,y,'o:b',ms = 9)
plt.show()
```

```python
#mec - marker edge color
x = np.array([1,2,3,4])
y = np.array([5,1,7,3])
plt.plot(x,y,'o:b',mec = 'r')
plt.show()
```

```python
#mfc - marker face color
x = np.array([1,2,3,4])
y = np.array([5,1,7,3])
plt.plot(x,y,'o:b',mfc = 'g')
plt.show()
```

```python
#linestyle shortly ls
#linestyles solid '-',none-'or',dotted-':'
y = np.array([3,7,1,9])
plt.plot(y, linestyle='dashed')#if we don't mentioned x,it will take 0,1,2,3....
plt.show()
```

```python
#line color shortly c
plt.plot(y,c='hotpink')
plt.show()
```

```python
#line width shortly lw
plt.plot(y,c='k',lw = '10')
plt.show()
```

```python
#multiple lines
y1 = np.array([2,10,4,9])
y2 = np.array([10,2,9,4])
plt.plot(y1,c='b')
plt.plot(y2,c='g')
plt.show()
```

# Labels and Titles

```python
x = np.array([2,6,8,10])
y = np.array([1,3,4,5])
plt.plot(x,y)
plt.xlabel("cost")
plt.ylabel("#choclates")
plt.title("Cost vs Choclate")
plt.show()
```

```python
#fonts of labels & titles
x = np.array([2,6,8,10])
y = np.array([1,3,4,5])
plt.plot(x,y)
font1 = {'family':'serif','color':'blue','size':10}
font2 = {'family':'serif','color':'green','size':20}
plt.xlabel("cost",fontdict=font1)
plt.ylabel("#choclates",fontdict=font1)
plt.title("Cost vs Choclate",fontdict=font2)
plt.show()
```

```python
#loaction of title
x = np.array([2,6,8,10])
y = np.array([1,3,4,5])
plt.plot(x,y)
plt.xlabel("cost")
plt.ylabel("#choclates")
plt.title("Cost vs Choclate",loc='right')
plt.show()
```

# subplot
plotting multiple plots in one area

```python
x=np.array([5,10,5,10,5,10])
plt.subplot(1,2,1)#(#row,#col,i)
plt.plot(x)
plt.title("1")

y=np.array([1,10,6,9])
plt.subplot(1,2,2)
plt.plot(y)
plt.title("2")

plt.suptitle("Plots")
plt.show()
```

```python
# hight dimension array of subplot

x1 =np.random.randint(20, size=(4))#range,#elements
plt.subplot(2,2,1)
plt.plot(x1,marker='o')

x2 =np.random.randint(2, size=(3))#range,#elements
plt.subplot(2,2,2)
plt.plot(x2,marker='o')

x3 = np.random.randint(10,size=(8))
plt.subplot(2,2,3)
plt.plot(x3,marker='o')

x4 = np.random.randint(12,size=(10))
plt.subplot(2,2,4)
plt.plot(x4,marker='o')

plt.show()
```

# Scatter

```python
#plotting scatter requires two data sets and size of then should be same
x = np.random.randint(10,size=(20))
y = np.random.randint(10,size=(20))
plt.scatter(x,y,color='hotpink')
plt.show()
```

```python
x = np.random.randint(10,size=(20))
y = np.random.randint(10,size=(20))
plt.scatter(x,y,color='hotpink')

x = np.random.randint(10,size=(20))
y = np.random.randint(10,size=(20))
plt.scatter(x,y,color='k')
plt.show()
```

```python
x = np.random.randint(10,size=(20))
y = np.random.randint(15,size=(20))
sizes = np.random.randint(100,size=(20))
plt.scatter(x,y,s=sizes)
plt.show()
```
